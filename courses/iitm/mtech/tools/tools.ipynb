### Section deals with Matrix manipulations 

```python
# Generate matrix transpose
import numpy as np

array = np.array([[1,1,0],[0,1,1]])

print(array.transpose())
```

```python
array = np.array([[-2,8,20],[14,19,10],[2,-2,1]])

print(array.transpose())
```

```python
# Matrix multiplication
array = np.array([[1,1,0],[0,1,1]])

print(np.matmul(array.transpose(), array))
print(np.matmul(array, array.transpose()))
```

```python
array = np.array([[-2,14,2],[8,19,-2],[20,10,1]])

print(np.matmul(array.transpose(), array))
print(np.matmul(array, array.transpose()))
```

### This section tool takes care of SVD related mathematical operations

```python
# Load required libraries
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
```

```python
array = np.array([[4,1,0],[1,3,1],[0,1,2]])

U, S, VT = np.linalg.svd(array, full_matrices=False)

print(U)

print(np.diag(S))

print(VT)
```
