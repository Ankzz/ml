```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Setup Data (Replace this with your actual data loading)
np.random.seed(42)
true_mean = 30
known_std = 8
# Generating 10 days of data, 20 samples each
data = np.random.normal(true_mean, known_std, size=(10, 20))

# 2. Initialization (Day 1 Prior)
day1_samples = data[0]
prior_mean = np.mean(day1_samples)
prior_var = np.var(day1_samples)
n_per_day = 20
sigma_sq = known_std**2

# Arrays to store updates for plotting
means = [prior_mean]
vars = [prior_var]

# 3. Recursive Bayesian Update
print(f"Initial Prior (from Day 1): Mean={prior_mean:.2f}, Var={prior_var:.2f}")

# We iterate from day 1 to day 10 (using Day 1 to update itself as per instructions)
for i in range(10):
    sample_mean = np.mean(data[i])
    
    # Precision Update: 1/post_var = 1/prior_var + n/sigma_sq
    new_precision = (1 / vars[-1]) + (n_per_day / sigma_sq)
    new_var = 1 / new_precision
    
    # Mean Update: Weighted average based on precision
    new_mean = ( (1 / vars[-1]) * means[-1] + (n_per_day / sigma_sq) * sample_mean ) / new_precision
    
    means.append(new_mean)
    vars.append(new_var)
    print(f"Day {i+1} Update: Posterior Mean={new_mean:.2f}, Posterior Var={new_var:.2f}")

# 4. Plotting the Distributions
plt.figure(figsize=(10, 6))
x = np.linspace(min(means)-10, max(means)+10, 500)

for i in range(len(means)):
    label = f'Prior (Day 1)' if i == 0 else f'Day {i} Post.'
    y = norm.pdf(x, means[i], np.sqrt(vars[i]))
    plt.plot(x, y, label=label, alpha=0.7 if i < 10 else 1.0, lw=2 if i == 10 else 1)

plt.title('Evolution of Posterior Distribution for Oncogene Mean Expression')
plt.xlabel('Mean Expression Level ($\mu$)')
plt.ylabel('Probability Density')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
