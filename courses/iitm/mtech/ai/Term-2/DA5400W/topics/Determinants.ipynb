### Determinant of a Matrix  

The **determinant** is a scalar value computed from a square matrix. It encodes important properties such as whether the matrix is invertible, the scaling factor of linear transformations, and the volume distortion in geometry.  

---

#### 📐 Definition  
For a square matrix \(A\), the determinant is denoted as \(\det(A)\) or \(|A|\).  

---

#### 🔎 Examples  

**1. 2×2 Matrix**  
\[
$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$
\]  

\[
$\det(A) = ad - bc$
\]  

---

**2. 3×3 Matrix**  
\[
$A = \begin{bmatrix} 
a & b & c \\ 
d & e & f \\ 
g & h & i 
\end{bmatrix}$
\]  

\[
$\det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)$
\]  

---

#### ⚙️ Properties of Determinants  
- \($\det(A) = 0$\) → matrix is **singular** (non-invertible).  
- \($\det(AB) = \det(A)\det(B)$\).  
- \($\det(A^T) = \det(A)$\).  
- Swapping two rows (or columns) changes the sign of the determinant.  
- If a matrix has a row or column of zeros, its determinant is 0.  
- Geometric meaning: determinant = scaling factor of the linear transformation defined by the matrix.  

---

#### ✅ Example Calculation  
\[
$A = \begin{bmatrix} 2 & 3 \\ 1 & 4 \end{bmatrix}$
\]  

\[
$\det(A) = (2)(4) - (3)(1) = 8 - 3 = 5$
\]  

So this matrix scales areas by a factor of 5.
