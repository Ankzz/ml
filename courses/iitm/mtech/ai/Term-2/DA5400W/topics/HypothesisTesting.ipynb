### 📊 Hypothesis Testing — Explained  

**Hypothesis testing** is a statistical method used to make decisions about population parameters based on sample data. It helps us determine whether observed results are due to chance or reflect a real effect.  

---

#### 🔑 Core Concepts
- **Null Hypothesis ($H_0$)**: The default assumption (e.g., “no effect,” “no difference”).  
- **Alternative Hypothesis ($H_1$)**: What we want to test (e.g., “there is an effect,” “means differ”).  
- **Test Statistic**: A value computed from sample data that measures how far the data deviates from $H_0$.  
- **p-value**: Probability of observing results as extreme as the sample, assuming $H_0$ is true.  
- **Significance Level ($\alpha$)**: Threshold (commonly 0.05) for deciding whether to reject $H_0$.  

---

#### 📐 Example 1: One-Sample Mean Test
Suppose we want to test if the average height of a population is 170 cm.  
- $H_0: \mu = 170$  
- $H_1: \mu \neq 170$  
- Collect sample data, compute sample mean $\bar{x}$, and test statistic:  
  $
  t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}
  $  
- Compare $t$ to critical values or compute p-value.  

---

#### 📐 Example 2: Coin Toss
We toss a coin 100 times, get 60 heads.  
- $H_0: p = 0.5$ (fair coin)  
- $H_1: p \neq 0.5$  
- Test statistic (z-test for proportions):  
  $
  z = \frac{\hat{p} - 0.5}{\sqrt{0.5(1-0.5)/n}}
  $  
  where $\hat{p} = 0.6$.  
- If $z$ is too large in magnitude, reject $H_0$.  

---

#### ⚖️ Types of Tests
- **Z-test / t-test:** Compare means.  
- **Chi-square test:** Compare categorical frequencies.  
- **ANOVA:** Compare multiple group means.  
- **Non-parametric tests:** For data not meeting normality assumptions.  

---

#### ✅ Applications
- Medicine: Testing effectiveness of treatments.  
- Business: A/B testing for marketing strategies.  
- Engineering: Quality control.  
- Machine Learning: Model comparison.
