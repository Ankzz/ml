### 📐 Parameter Estimation — Explained  

**Parameter estimation** is the process of using data to infer the values of unknown parameters in a mathematical model. It’s a central concept in statistics, machine learning, and optimization.  

---

#### 🔑 Two Main Approaches
1. **Point Estimation**  
   - Produces a single “best guess” of the parameter.  
   - Examples:  
     - **Sample mean** $\hat{\mu} = \frac{1}{n}\sum x_i$ estimates the population mean.  
     - **Sample variance** estimates the population variance.  

2. **Interval Estimation**  
   - Produces a range of plausible values (confidence interval).  
   - Example: A 95% confidence interval for the mean might be $[5.2, 6.8]$.  

---

#### 📊 Common Methods
- **Maximum Likelihood Estimation (MLE):**  
  Choose parameters that maximize the likelihood of observing the given data.  
  Example: For normal distribution, MLE for mean = sample mean.  

- **Method of Moments:**  
  Match sample moments (mean, variance, etc.) with theoretical moments.  

- **Bayesian Estimation:**  
  Treat parameters as random variables with prior distributions, update beliefs using Bayes’ theorem.  

---

#### ✅ Example
Suppose we assume data comes from a normal distribution $N(\mu, \sigma^2)$.  
- Given sample: $[5, 6, 7]$.  
- Sample mean = $\hat{\mu} = \frac{5+6+7}{3} = 6$.  
- Sample variance = $\hat{\sigma}^2 = \frac{(5-6)^2 + (6-6)^2 + (7-6)^2}{3} = \frac{2}{3}$.  

So estimated parameters:  
$
\mu \approx 6, \quad \sigma^2 \approx 0.67
$

---

#### ⚖️ Applications
- **Statistics:** Estimating population parameters.  
- **Machine Learning:** Training models (weights are parameters).  
- **Economics:** Estimating demand/supply curves.  
- **Engineering:** System identification and control.

### 📐 Maximum Likelihood Estimation (MLE)  

**Maximum Likelihood Estimation (MLE)** is a method of estimating parameters of a statistical model by choosing values that maximize the likelihood of observing the given data.  

---

#### 🔑 Core Idea
- Suppose data points $x_1, x_2, \dots, x_n$ come from a distribution with parameter $\theta$.  
- The **likelihood function** is:  
  $
  L(\theta) = \prod_{i=1}^n f(x_i \mid \theta)
  $  
  where $f(x_i \mid \theta)$ is the probability density (or mass) function.  
- The MLE $\hat{\theta}$ is the value of $\theta$ that maximizes $L(\theta)$.  
- Often we maximize the **log-likelihood**:  
  $
  \ell(\theta) = \ln L(\theta) = \sum_{i=1}^n \ln f(x_i \mid \theta)
  $  
  because logs turn products into sums and simplify differentiation.  

---

#### ✅ Example 1: Bernoulli Distribution
Data: coin flips $x_i \in \{0,1\}$.  
Parameter: probability of heads $p$.  
Likelihood:  
$
L(p) = \prod_{i=1}^n p^{x_i}(1-p)^{1-x_i}
$  
Log-likelihood:  
$
\ell(p) = \sum_{i=1}^n \big[x_i \ln p + (1-x_i)\ln(1-p)\big]
$  
Differentiate and solve:  
$
\hat{p} = \frac{\sum x_i}{n}
$  
So the MLE for $p$ is just the **sample mean**.  

---

#### ✅ Example 2: Normal Distribution
Data: $x_1, \dots, x_n \sim N(\mu, \sigma^2)$.  
Parameters: $\mu, \sigma^2$.  
Log-likelihood:  
$
\ell(\mu, \sigma^2) = -\frac{n}{2}\ln(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^n (x_i - \mu)^2
$  
Differentiate:  
- MLE for mean: $\hat{\mu} = \frac{1}{n}\sum x_i$ (sample mean).  
- MLE for variance: $\hat{\sigma}^2 = \frac{1}{n}\sum (x_i - \hat{\mu})^2$ (sample variance, note denominator $n$ not $n-1$).  

---

#### ⚖️ Applications
- **Statistics:** Estimating distribution parameters.  
- **Machine Learning:** Training models (logistic regression, neural networks).  
- **Econometrics:** Estimating demand/supply models.  
- **Signal Processing:** Noise parameter estimation.

### 📐 Method of Moments (MoM)  

The **Method of Moments** is another classical technique for parameter estimation. Instead of maximizing likelihoods (like MLE), it matches **sample moments** (mean, variance, etc.) with the corresponding **theoretical moments** of the distribution.  

---

#### 🔑 Core Idea
- A distribution has theoretical moments (expectations of powers of the random variable).  
  - First moment: $E[X]$ → mean  
  - Second moment: $E[X^2]$ → relates to variance  
- The sample has empirical moments:  
  - First sample moment: $\hat{m}_1 = \frac{1}{n}\sum x_i$  
  - Second sample moment: $\hat{m}_2 = \frac{1}{n}\sum x_i^2$  
- The method sets:  
  $
  \hat{m}_k = \text{theoretical moment of order } k
  $  
  and solves for parameters.  

---

#### ✅ Example 1: Exponential Distribution
Suppose $X \sim \text{Exp}(\lambda)$.  
- Theoretical mean: $E[X] = \frac{1}{\lambda}$.  
- Sample mean: $\hat{m}_1 = \frac{1}{n}\sum x_i$.  
Set equal:  
$
\hat{m}_1 = \frac{1}{\lambda} \quad \implies \quad \hat{\lambda} = \frac{1}{\hat{m}_1}
$  

So the MoM estimator for $\lambda$ is the reciprocal of the sample mean.  

---

#### ✅ Example 2: Normal Distribution
Suppose $X \sim N(\mu, \sigma^2)$.  
- Theoretical mean: $E[X] = \mu$.  
- Theoretical second moment: $E[X^2] = \mu^2 + \sigma^2$.  
- Sample mean: $\hat{m}_1 = \frac{1}{n}\sum x_i$.  
- Sample second moment: $\hat{m}_2 = \frac{1}{n}\sum x_i^2$.  

Set equal:  
$
\hat{m}_1 = \mu, \quad \hat{m}_2 = \mu^2 + \sigma^2
$  

Solve:  
$
\hat{\mu} = \hat{m}_1, \quad \hat{\sigma}^2 = \hat{m}_2 - \hat{m}_1^2
$  

---

#### ⚖️ Comparison with MLE
- **MoM:** Simple, intuitive, often easier to compute.  
- **MLE:** Statistically more efficient (lower variance, asymptotically unbiased).  
- **MoM:** May give biased estimates but is useful when likelihoods are hard to compute.

### 📐 Bayesian Estimation  

**Bayesian estimation** is a statistical method that treats parameters as random variables with probability distributions, rather than fixed but unknown values. It combines prior beliefs with observed data to produce updated (posterior) beliefs.  

---

#### 🔑 Core Idea
Bayes’ theorem is the foundation:

$
P(\theta \mid \text{data}) = \frac{P(\text{data} \mid \theta) \cdot P(\theta)}{P(\text{data})}
$

- **Prior $P(\theta)$:** What we believe about the parameter before seeing data.  
- **Likelihood $P(\text{data} \mid \theta)$:** How probable the observed data is given the parameter.  
- **Posterior $P(\theta \mid \text{data})$:** Updated belief after seeing data.  
- **Evidence $P(\text{data})$:** Normalizing constant ensuring probabilities sum to 1.  

---

#### ✅ Example 1: Coin Toss
Suppose we want to estimate the probability of heads $p$.  
- Prior: $p \sim \text{Beta}(\alpha, \beta)$.  
- Likelihood: $n$ tosses, $k$ heads → $p^k (1-p)^{n-k}$.  
- Posterior:  
  $
  p \mid \text{data} \sim \text{Beta}(\alpha + k, \beta + (n-k))
  $  

So Bayesian estimation updates the prior parameters with observed counts.  

---

#### ✅ Example 2: Normal Distribution Mean
Suppose data $x_1, \dots, x_n \sim N(\mu, \sigma^2)$, with known $\sigma^2$.  
- Prior: $\mu \sim N(\mu_0, \tau^2)$.  
- Likelihood: Normal with mean $\mu$.  
- Posterior: $\mu \mid \text{data}$ is also normal, with mean a weighted average of prior mean and sample mean.  

This shows how Bayesian estimation balances **prior knowledge** with **observed evidence**.  

---

#### ⚖️ Comparison
- **MLE:** Chooses parameter that maximizes likelihood.  
- **MoM:** Matches sample moments to theoretical ones.  
- **Bayesian Estimation:** Produces a full posterior distribution, not just a point estimate.  

---

#### 📊 Applications
- Machine learning (Bayesian neural networks, probabilistic models).  
- Economics (updating beliefs about market parameters).  
- Medicine (diagnostic probabilities).  
- Engineering (system reliability estimation).
