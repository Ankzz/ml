### Jordan Canonical Form (Jordan Normal Form)

The **Jordan form** is a way of representing a square matrix in a nearly diagonal structure that reveals its eigenvalues and the structure of its eigenspaces. It’s especially useful when a matrix is **not diagonalizable**.

---

#### 🔑 Key Idea
- Every square matrix $A$ can be transformed (via similarity transformation) into a **Jordan matrix** $J$:  
  $
  $A = PJP^{-1}$
  $  
  where $J$ is block diagonal, and each block is a **Jordan block**.

---

#### 📐 Jordan Block
A Jordan block associated with eigenvalue $\lambda$ looks like:
$
$J_k(\lambda) =
\begin{bmatrix}
\lambda & 1 & 0 & \cdots & 0 \\
0 & \lambda & 1 & \cdots & 0 \\
0 & 0 & \lambda & \cdots & 0 \\
\vdots & \vdots & \vdots & \ddots & 1 \\
0 & 0 & 0 & \cdots & \lambda
\end{bmatrix}_{k \times k}$
$

- Diagonal entries = eigenvalue $\lambda$.  
- Superdiagonal entries = 1.  
- All other entries = 0.  

---

#### ⚖️ Why Jordan Form Matters
- **Diagonalizable case:** If every eigenvalue’s geometric multiplicity = algebraic multiplicity, then all Jordan blocks are $1 \times 1$. The Jordan form is simply a diagonal matrix of eigenvalues.  
- **Defective case:** If some eigenvalues don’t have enough independent eigenvectors, Jordan blocks larger than $1 \times 1$ appear.  

---

#### ✅ Example
Matrix:
$
$A = \begin{bmatrix}
2 & 1 \\
0 & 2
\end{bmatrix}$
$

- Eigenvalue: $\lambda = 2$, algebraic multiplicity = 2, geometric multiplicity = 1.  
- Not diagonalizable.  
- Jordan form:
$
$J = \begin{bmatrix}
2 & 1 \\
0 & 2
\end{bmatrix}$
$
(one Jordan block of size 2).  

---

#### 🔎 Applications
- **Differential equations:** Jordan form simplifies solving systems of linear ODEs.  
- **Matrix functions:** Computing $e^A$, $\sin(A)$, etc., is easier in Jordan form.  
- **Theoretical insight:** Shows the structure of linear transformations beyond diagonalization.
