```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# 1D dataset
X = np.array([1, 2, 5, 8, 9]).reshape(-1, 1)

# Perform hierarchical clustering (single linkage)
Z = linkage(X, method='average')

# Show the linkage matrix (merging steps)
print("Linkage matrix:\n", Z)

# Plot dendrogram
plt.figure(figsize=(6, 4))
dendrogram(Z, labels=[1, 2, 5, 8, 9])
plt.title("Agglomerative Clustering (1D)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```
