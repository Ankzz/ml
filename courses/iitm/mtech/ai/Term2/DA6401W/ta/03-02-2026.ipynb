# Quadractic Programming Problems

**Problem 1**: Distance Minimization (Machine Learning)Problem StatementIn Support Vector Machines (SVM), the goal is to find a "hyperplane" that separates two classes of data while maximizing the "margin" between them. Mathematically, this is equivalent to minimizing the squared norm of the weight vector.Minimize:$$f(\mathbf{w}) = \frac{1}{2} \mathbf{w}^T \mathbf{w}$$Subject to:$y_i(\mathbf{w}^T \mathbf{x}_i + b) \geq 1$ (Ensuring all points are correctly classified outside the margin)Question: If you have two points $\mathbf{x}_1 = [1, 2]$ and $\mathbf{x}_2 = [3, 4]$, find the vector $\mathbf{w}$ that minimizes the "stress" (length) while keeping the points separated.

**Problem 2**: Heat Transfer & Energy (Physics/Engineering)Problem StatementIn structural engineering or heat transfer, the Principle of Minimum Potential Energy is often used. Imagine a system of three connected springs. The "Energy" in the system is a quadratic function of the displacement of the nodes ($x_1, x_2$).Minimize (Total Potential Energy):$$E(x_1, x_2) = \frac{1}{2} [x_1, x_2] \begin{bmatrix} 10 & -5 \\ -5 & 10 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} - \begin{bmatrix} 2 \\ 0 \end{bmatrix}^T \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$$Subject to:$x_1 + x_2 \leq 5$ (Physical barrier constraint)$x_1, x_2 \geq 0$Question: Find the equilibrium displacements $x_1$ and $x_2$ that minimize the total energy.

**Problem 3**: Production Smoothing (Supply Chain)Problem StatementA factory wants to plan production ($x_t$) for the next two months. Changing production levels quickly is expensive (wear and tear, hiring/firing costs), so they want to minimize the squared change in production.Minimize (Smoothness Cost):$$f(x_1, x_2) = (x_1 - 50)^2 + (x_2 - x_1)^2$$(Where 50 was the production level last month)Subject to:$x_1 + x_2 \geq 120$ (Total demand for two months)$x_t \leq 80$ (Maximum machine capacity per month)Question: What production schedule ($x_1, x_2$) minimizes the "shock" to the system while meeting total demand?
