### Algebraic vs. Geometric Multiplicity  

When we solve for eigenvalues of a matrix, two important notions arise: **algebraic multiplicity** and **geometric multiplicity**.  

---

#### 📐 Algebraic Multiplicity (AM)  
- **Definition:** The number of times an eigenvalue appears as a root of the characteristic polynomial.  
- **Example:** If the characteristic polynomial is \((\lambda - 2)^3(\lambda - 5)\), then:  
  - Eigenvalue \(2\) has algebraic multiplicity = 3  
  - Eigenvalue \(5\) has algebraic multiplicity = 1  

---

#### 📐 Geometric Multiplicity (GM)  
- **Definition:** The dimension of the eigenspace corresponding to an eigenvalue (i.e., the number of linearly independent eigenvectors associated with that eigenvalue).  
- **Example:** If eigenvalue \(2\) has 3 as algebraic multiplicity but only 2 independent eigenvectors, then:  
  - Geometric multiplicity = 2  

---

#### 🔎 Relationship Between AM and GM  
- Always:  
  \[
  $1 \leq \text{GM}(\lambda) \leq \text{AM}(\lambda)$
  \]  
- If GM = AM for all eigenvalues → the matrix is **diagonalizable**.  
- If GM < AM for some eigenvalue → the matrix is **defective** (not diagonalizable).  

---

#### ✅ Example  
Matrix:  
\[
$A = \begin{bmatrix} 2 & 1 \\ 0 & 2 \end{bmatrix}$
\]  

- Characteristic polynomial: \((\lambda - 2)^2\) → eigenvalue \(2\) has **algebraic multiplicity = 2**.  
- Solve \((A - 2I)x = 0\):  
\[
$\begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix} x = 0$
\]  
This gives only **one independent eigenvector** → **geometric multiplicity = 1**.  

So here:  
- AM = 2  
- GM = 1  
- Matrix is **not diagonalizable**.
