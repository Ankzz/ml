# Decision Tree 

## First approach just demonstrates decision tree using Iris dataset available 

```python
# Step 1: Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Step 2: Load dataset (Iris dataset as example)
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Step 3: Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

X_test
```

```python
# Step 4: Initialize and train the decision tree
clf = DecisionTreeClassifier(
    criterion="gini",   # or "entropy"
    max_depth=3,        # limit depth to avoid overfitting
    random_state=42
)
clf.fit(X_train, y_train)
```

```python
# Step 5: Make predictions
y_pred = clf.predict(X_test)

# Step 6: Evaluate accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Step 2: Encode categorical variables
from sklearn.preprocessing import LabelEncoder

class DestinationLabelEncoder(LabelEncoder):
    def fit_transform(self, y):
        # print(f"Call lands here: {y}")
        if 'Other' not in y.values:
            #o_series = pd.Series([])
            y[max(y.index)+1] = 'Other' 
        # print(f"Updated: {y}")
        fit = super().fit_transform(y)
        return fit

    def transform(self, y):
        if y not in self.classes_:
            y = ['Other']
        return super().transform(y)

dest_encoder = DestinationLabelEncoder()
label_encoder = LabelEncoder()
initial_size = len(df)
X = dest_encoder.fit_transform(df["destination"]).reshape(-1, 1)
if initial_size != len(X):
    print("Adding another label")
    new_entry = pd.DataFrame({"label": ["Restricted travel - Visa Required"]})
    df = pd.concat([df, new_entry], ignore_index=True)
    # print(df["label"])
y = label_encoder.fit_transform(df["label"])

print(f"Data now ready for training!")
```

```python
# Step 3: Train decision tree
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X, y)

print("Model Training completed!")
```

```python
# Step 4: Predict
test_destination = "India"
encoded_test = dest_encoder.transform([test_destination]).reshape(-1, 1)
pred = clf.predict(encoded_test)

print("Prediction:", label_encoder.inverse_transform(pred)[0])
```
