### Problem 1

#### String Operations
Write a function named string_operations(s: str) that accepts a string and returns a tuple: (vowel count: int, replaced: str, is palindrome: bool).

The function must:

Count the number of vowels (a,e,i,o,u — count both upper and lower case).
Replace all spaces with underscores and return the replaced string as the second element.
Check if the string is a palindrome and return the boolean as the third element. Definition used: ignore case and ignore spaces when checking palindrome (i.e.,treat only alphanumeric characters and ignore case).

```python
def count_vowels(s: str):
    vowels = ['a','e','i','o','u', 'A', 'E', 'I', 'O', 'U']
    count = 0
    for x in s:
        if x in vowels:
            count = count + 1
    return count

def replace_spaces(s: str):
    r = ""
    for x in s:
        if x == " ":
            r = r + "_"
        else:
            r = r + x
    return r            

def is_pallindrome(s: str):
    i=0
    j=len(s)-1
    s = s.lower()
    while i<j:
        if s[i] != s[j]:
            if s[i]==" ":
                i = i+1
                continue
            if s[j]==" ":
                j = j-1
                continue
            break
        i = i + 1
        j = j - 1

    return i>=j    

import re
def string_operations(s: str):
    '''
    Count vowels, replace spaces with underscores, 
    and check if string is a palindrome (ignore case and non-alphanumeric).
    '''
    # write your code here
    if s == None or s =="":
        return (0,s, True)
    else:
        vowel_count = count_vowels(s)
        r = replace_spaces(s)
        s = re.sub(r'[^A-Za-z0-9]', '', s).lower()
        pallindrome = is_pallindrome(s)
        return (vowel_count, r, pallindrome)

# print(string_operations(None))
print(string_operations(""))
print(string_operations("None"))
print(string_operations("No one"))
print(string_operations("No on"))
print(string_operations("Race car"))
print(string_operations("Race car "))
print(string_operations(" Race car "))
print(string_operations("n n "))
print(string_operations("a"))
print(string_operations("A "))
print(string_operations(" "))
```

### Problem 2

__Factorial__

Implement two functions: factorial_rec(n: int) -> int (recursive) and factorial_iter(n: int) -> int (iterative). Compare results for both both functions, it must return the same integer value.

```python
def factorial(n: int, M:{}) -> int:
    if n in M:
        return M[n]
    M[n] = n * factorial(n-1, M)

    return M[n]

def factorial_rec(n: int) -> int:
    if n<0:
        return -1
    M = {0:1,1:1,2:2,3:6}
    return factorial(n, M)
    
def factorial_iter(n: int) -> int:
    if n<0:
        return -1
    if n == 0 or n==1:
        return 1    
    f = 1
    while n>0:
        f = f * n
        n = n -1

    return f

print(factorial_rec(5))
print(factorial_iter(5))
print(factorial_rec(10))
print(factorial_iter(10))    
```

__Filter average__
Given a list of tuples where each tuple is (name, list of scores), implement a function filter_avg(data: list) -> dict using list comprehensions and no loops.

The resulting dictionary should map names to the average of scores that are >= 75. If a person has no scores >= 75, exclude them from the dictionary.

Note: The solution should not have any for loops or while loops

Example

    Input:

    data = [
        ("Alice",[80,70,90]),
        ("Bob",[60,50,75]),
        ("Charlie",[90,85,92]),
        ("David",[45,65,70])
    ]
    Output:

    {'Alice': 85.0, 'Bob': 75.0, 'Charlie': 89.0}

```python
def is_filtered(elem):
    data = list(filter(lambda x: x>=75, elem[1]))
    avg = 0
    if len(data)>=1:
        avg = sum(data)/len(data)
    return avg>=75

def get_avg(elem):
    data = list(filter(lambda x: x>=75, elem))
    avg = sum(data)/len(data)
    return avg

def filter_avg(data: list) -> dict:
    data = list(filter(is_filtered, data))
    data = [(x[0],get_avg(x[1])) for x in data]
    return dict(data)

data = [
        ("Alice",[80,70,90]),
        ("Bob",[60,50,78]),
        ("Charlie",[90,85,92]),
        ("David",[45,65,70])
    ]

print(filter_avg((data)))    
```

### Problem 7

Write a function find_top_students(students: list) -> dict where each student is represented as a dictionary: "name": str, "scores": list[int]. The function should:

Compute the average score for each student.
Filter students with average ≥ 85.
Return a new dictionary mapping student name to their average score rounded to one decimal place (use round(avg,1)).
Example

Input:

students = [

   {"name":"Bob","scores":[70,60,75]},

Build a Library system using classes (exact names required):

Base class: LibraryUser with attributes name and user_id. Implement __str__() returning: User: <name> (ID: <user id>).

Subclasses:

    StudentUser(LibraryUser) — adds attribute grade_level. Override __str__() to return: Student: <name> (ID: <user id>), 

    Grade: <grade level>. Implement borrow_book(title: str) -> str that returns: "<name> borrowed '<title>' for 14 days.

    FacultyUser(LibraryUser) — adds attribute department. Override __str__() to return: "Faculty: <name> (ID: <user id>), 

    Dept: <department>". Implement borrow_book(title: str) -> str that returns: "<name> borrowed '<title>' for 30 days."

    Example

    Code

    u = StudentUser('Alice','S123','10th')
    print(u)
    print(u.borrow_book('Python Basics'))
    Output

    Student: Alice (ID: S123), Grade: 10th
    Alice borrowed 'Python Basics' for 14 days.




```python
class LibraryUser:
    def __init__(self, name: str, user_id: str):
        self.name = name
        self.user_id = user_id
        
    def __str__(self):
        return f"User: {self.name} (ID: {self.user_id})"
        

class StudentUser(LibraryUser):
    def __init__(self, name: str, user_id: str, grade_level: str):
        super().__init__(name, user_id)
        self.grade_level = grade_level
    
    def borrow_book(self, title: str) -> str:
        return f"{self.name} borrowed '{title}' for 14 days."

    def __str__(self):
        return f"Student: {self.name} (ID: {self.user_id}), Grade: {self.grade_level}"


class FacultyUser(LibraryUser):
    def __init__(self, name: str, user_id: str, department: str):
        super().__init__(name, user_id)
        self.department = department
    
    def borrow_book(self, title: str) -> str:
        return f"{self.name} borrowed '{title}' for 30 days."
    
    def __str__(self):
        return f"Faculty: {self.name} (ID: {self.user_id}), Dept: {self.department}"


u = StudentUser("Alice","S123","10th")
print(u)
print(u.borrow_book("Python Basics"))        
u = FacultyUser("Bob","F123","Stats")
print(u)
print(u.borrow_book("Python Basics"))      
```

### Problem 10

__Course Registration System__

Course Registration system (exact names required):

Class Course with attributes: course_id: str, title: str, credits: int.

Class Student with attribute name: str and a list to hold registered Course objects. Methods:

register_for_course(self, course: Course) -> None

calculate_total_credits(self) -> int
"__str__()" -> str - display student’s name, registered courses with credits and total credits (see example below).

    Example:

    c1 = Course("CS101", "Intro to CS", 3)
    c2 = Course("MA101", "Calculus I", 4)
    s1 = Student("Alice")
    s1.register_for_course(c1)
    s1.register_for_course(c2)
    print(s1)
    Expected Output:

    Student: Alice
    Registered Courses:
    CS101 - Intro to CS (Credits: 3)
    MA101 - Calculus I (Credits: 4)
    Total Credits: 7

```python
class Course:
    
    def __init__(self, course_id: str, title: str, credits: int):
        self.course_id = course_id
        self.title = title
        self.credits = credits

    def __str__(self):
        return f"Course({self.course_id}, {self.title}, {self.credits})"

class Student:
    
    def __init__(self, name: str):
        self.name = name
        self.courses = []
    
    def register_for_course(self, course: Course) -> None:
        self.courses.append(course)
    
    def calculate_total_credits(self) -> int:
        return sum(c.credits for c in self.courses)
    
    def __str__(self) -> str:
        lines = [f"Student: {self.name}", "Registered Courses:"]
        for c in self.courses:
            lines.append(f"{c.course_id} - {c.title} (Credits: {c.credits})")
        lines.append(f"Total Credits: {self.calculate_total_credits()}")
        return "\n".join(lines)
    
c1 = Course("CS101", "Intro to CS", 3)
c2 = Course("MA101", "Calculus I", 4)
s1 = Student("Alice")
s1.register_for_course(c1)
s1.register_for_course(c2)
print(s1)
```
