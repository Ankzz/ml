Question 3: Employee Satisfaction and Performance Analysis

HR Department wants to analyze employee satisfaction, performance metrics, and department-wise patterns to improve workplace policies and retention.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
# Create Dataset
np.random.seed(45)

n_employees = 150
employee_data = pd.DataFrame({
    'Department': np.random.choice(['Sales', 'Engineering', 'Marketing', 'HR', 'Finance'], n_employees),
    'Years_Experience': np.random.randint(1, 15, n_employees),
    'Satisfaction_Score': np.random.uniform(3,10, n_employees),
    'Perfromance_Score': np.random.uniform(50, 95, n_employees),
    'Training_Hours': np.random.randint(10,100, n_employees),
    'Salary': np.random.uniform(40000, 120000, n_employees),
    'Work_Life_Balance': np.random.uniform(3, 10, n_employees)
})

# Adding some co-relation
employee_data['Satisfaction_Score'] = (employee_data['Satisfaction_Score']  \
                                      + (employee_data['Work_Life_Balance']) * 0.3 \
                                      + (employee_data['Training_Hours'] * 0.02) )
```
