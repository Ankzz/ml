## Sorting and Ranking Dataframes

```python
df = pd.DataFrame({
    "Score_A": [30,12,45,33],
    "Score_B": [100,101,102,90]
})

df['Rank_A_Avg'] = df['Score_A'].rank(method="average")
df
```

```python
df['Rank_B_Avg'] = df['Score_B'].rank(method="min")
df
```

## Concatenation of dataframes vertically

```python
# pd.concat
df1 = pd.DataFrame({
    "Key1": ["Val1", "Val2", "Val3", "Val4"]
    })

df2 = pd.DataFrame({
    "Key1": ["Val5", "Val6"]
    })

print(df1)
print(df2)
print(pd.concat([df1, df2]))
print(pd.concat([df1, df2], axis=1))
```
