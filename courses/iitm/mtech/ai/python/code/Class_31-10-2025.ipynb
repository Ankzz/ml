# Sorting in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("sample_data/SuperMarket Analysis.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df["Total"] = df["Sales"] + df["Tax 5%"]
```

```python
branch_sorted = df.groupby('Branch')
branch_sorted
```

```python
branch_sorted = df.sort_values(by="Total", ascending=False)
branch_sorted
```

```python
top_rated = df.sort_values(by="Rating")
top_rated
```

```python
df_index = df.set_index("Invoice ID")
df_index.rank(method='average')
```

```python
product_revenue = df.groupby("Product line")["Total"].sum().sort_values(ascending=False)
product_revenue
```

```python
product_revenue.rank(method='dense', ascending=False)
```

```python
product_revenue.rank(method='average', ascending=False)
```
