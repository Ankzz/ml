## Matplotlib Baiscs
* Works with lists, Numpy arrays, and Pandas Series
* Key Concepts:
  * figure: Overall plotting Canvas - for which dimensions can be specified
  * axes: Python object contains a plot placed within a figure (box/es on paper)
  * plt: Convenient handle for matplotlib objects - keeps track of the current figure and current axes automatically
  

## Matplotlib Demo

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.figure()
plt.plot(x, y, label="sin(x)")
plt.plot(x, z, label="cos(x)", linestyle="--")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
## Scatter Plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

plt.figure()
plt.scatter(x_scatter, y_scatter, color="red", marker="o")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Bar Plot
categories = ["A", "B", "C", "D"]
values = [5,7,2,9]

plt.figure()
plt.bar(categories, values, align='center')
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)

plt.figure()
plt.hist(data, bins=20)
plt.title("Histogram")
plt.ylabel("Frequency")
plt.xlabel("Value")
plt.show()
```

```python
# Subplots using Matplotlib
fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].plot(x, y, color="blue")
ax[0,0].set_title("sin(x)")

ax[0,1].plot(x, z, color="red")
ax[0,1].set_title("cos(x)")

ax[1,0].scatter(x_scatter, y_scatter, color="green")
ax[1,0].set_title("scatter")

ax[1,1].hist(data, bins=15, color="purple")
ax[1,1].set_title("histogram")

fig.suptitle("Subplots")

plt.show()
```

## Seaborn Basics

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load IRIS dataset ingrained in Seaborn
df = sns.load_dataset('iris')

df
```

```python
sns.set_theme(style="whitegrid", palette="Set2")
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    style="species",
    s=50
)
plt.title("Iris: Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df["petal_length"])
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df, x="species", y="petal_length")
plt.title("Iris: Species vs Sepal Length")
plt.ylabel("Sepal Length")
plt.xlabel("Species")
plt.show()
```

```python
plt.figure(figsize=(6, 4))

corr = df.select_dtypes("number").corr()
sns.heatmap(data=corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Coorelation Heatmap")
plt.show()
```

## Plotly Basics
* Interactive by default: Hover tool tips, zoom/pan, select, etc
* Can be rendered on browser and apps easily: uses JavaScript engine
* Provides high-level API for fast charts and low-level API for fine control
* Supports animations, subplots, 3D plots, maps, etc
* Core Concepts:
  * Figure: 
  * plotly.express
  * graph_objects

```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.iris()
df
```

```python
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species", symbol="species", hover_data=["sepal_length", "sepal_width"], title="Iris: Sepal Length vs Width")
fig.show()
```
