## Bisect

```python
import bisect
```

```python
my_list = [1,3,5,7]

index_right = bisect.bisect(my_list, 4)
print(index_right)
```

```python
index_left = bisect.bisect_left(my_list, 5)
print(index_left)
```

```python
print(my_list)
```

```python
# my_list = [7,3,5,1]
bisect.insort(my_list, 2)
print(my_list)
```

```python
print(my_list)
print(my_list[:-1])
print(my_list[-1:])
print(my_list[::-1])
```
