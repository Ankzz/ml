Q1. Generate Transpose of a Matrix

```python
matrix = [[1,2,3], [4,5,6]]

transpose = []

for i in range(len(matrix[0])):
    new_row = []
    for row in matrix:
        new_row.append(row[i])
    transpose.append(new_row)

print(matrix)
print(transpose)
```

Same code using List comprehension

```python
matrix =  [[1,2,3], [4,5,6]]
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]

print(transpose)
```

Q2: Dictionary of Word Lengths

```python
words = ["apple", "banana", "cherry"]

word_lengths = {w: len(w) for w in words}

print(word_lengths)
```

Q3: Filter Even Squares

```python
# Generate list of numbers from 1 to 20
nums = list(range(1,21))

square_list = [x*x for x in nums if x%2==0]

print(square_list)
```

Q4. Character Frequency

```python
def count_character_frequency(s:str) -> dict:
    d = {}
    d = {c:s.count(c) for c in set(s) if c.isalpha() }
    return d

print(count_character_frequency("practice makes a man / woman perfect"))
```

Q5. Swap keys & Values

```python
def swap_key_and_values(d:dict) -> dict:
    d = {v:k for k,v in d.items()}
    return d

print(swap_key_and_values({"key1": "value1", "key2": "value2"}))
print(swap_key_and_values({"key1": "value", "key2": "value"}))
```

Q6. Combine Lists into Dict

```python
names = ['A', 'B', 'C']
marks = [95, 85, 90]

d = {names[i]:marks[i] for i in range(len(names))}

print(d)
```

```python
d = {n:m for n,m in zip(names, marks)}

print(d)
```

Q7. Filter Dict by Condition

```python
scores = {'A': 95, 'B': 85, 'C': 90}
passed = {k:v for k,v in scores.items() if v>=90}

print(passed)
```

Q8. Flatten Nested List

```python
nested = [[1,2],[3,4]]

flat = [y for x in nested for y in x]

print(flat)
```

Q9. Dict from Pairs with Conditions

```python
pairs = [('a',1), ('b', 2), ('c', 3)]
d = {pair[0]:pair[1] for pair in pairs}
print(d)
d = {k:v for k,v in pairs}
print(d)
d = {pair[0]:pair[1] for pair in pairs if pair[1]>=2}
print(d)
```

Q10. Nested Comprehension for Table

```python
table = [[i*j for j in range(1,6)] for i in range(1,6)]

for row in table:
    print(row)
```

Q1. Distance2D - Compute distance between two points

```python
class Distance2D:
    def __init__(self, x1, y1, x2, y2):
        self.x1, self.y1, self.x2, self.y2 = float(x1), float(y1), float(x2), float(y2)

    def compute(self):
        return ((self.x1 - self.x2)**2 + (self.y1 - self.y2)**2)**0.5

p = Distance2D(1,2,3,4)
print(f"Distance ={round(p.compute(),2)}")        
```

Q2. CircleArea - Find area and circumference

```python
import math

class CircleArea:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius**2
    def circumference(self):
        return 2 * math.pi * self.radius

c = CircleArea(5)
print(f"Area: {round(c.area(),2)} Circumference: {round(c.circumference(), 2)}")
```

Q3, WordAnalyzer - Count vowels and consonants

```python
class WordAnalyzer:
    def __init__(self, text:str):
        self.text = text.lower()
    def vowels(self):
        return sum(1 for ch in self.text if ch in "aeiou")    
    def consonants(self):
        return sum(1 for ch in self.text if ch.isalpha() and ch not in "aeiou")
    
w = WordAnalyzer("hello ankit!")
print(f"Vowels: {w.vowels()} Consonants: {w.consonants()}")
```

Q4. SimpleCalculator - Perform basic operations

```python
class SimpleCalculator:
    def __init__(self, a, b):
        self.a, self.b = a,b
    def add(self): return (self.a+self.b)
    def subtract(self): return (self.a-self.b)
    def multiply(self): return (self.a * self.b)
    def divide(self): return (self.a / self.b) if self.b != 0 else None

calculator = SimpleCalculator(5, 10)
print(f"Add: {calculator.add()} Subtract: {calculator.subtract()} Multiply: {calculator.multiply()} Divide: {calculator.divide()}")
```

Q5. TemperatureConverter - Convert between Celsius and Fahrentheit

```python
class TemperatureConverter:
    def __init__(self, celsius):
        self.c = float(celsius)
    def to_fahrenheit(self):
        return (self.c * 9/5) + 32
    def to_kelvin(self):
        return self.c + 273.15
    
t = TemperatureConverter(30)
print(f"Celsius: {t.c} Fahrenheit: {t.to_fahrenheit()} Kelvin: {t.to_kelvin()}")
```
