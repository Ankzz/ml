# Lambda Functions
- A lambda is a short, one-line function without a name and referred to as anonymous function
- Defined using the keyword lambda, followed by parameters and expression
    Ex: lambda x: x**2
- Expression after : is automatically returned - no return keyword needed
- Commonly used to create a conditional rule without defining a full function
    Ex: nums = [3,5,8,10]
    list(filter(lambda x: x%2 == 0, nums))
- Can be used along with if/else condition and for loops as well
- Note: Ise lamda functions only for inline one-time transformations
- Widely used in feature creation, aggregation, and preprocessing steps of Machine Learning    

```python
# Traditional approach for function declaration
def square_(a: int):
    return a**2
print(square_(5))

# Lambda approach
square_lambda = lambda x: x**2
print(square_lambda(5))

# Lambda direct call
print((lambda x: x**2)(5))

# Lambda with multiple parameters
print((lambda x,y: x**2+y**2)(2,3))
```

## Lambda based filters

```python
nums = [3, 5, 8, 10]

list(filter(lambda x: x%2==0, nums))
```

```python
nums = [('a', 1), ('b', 2), ('c', 4)]
dict(filter(lambda x: x[1]%2==0, nums))
```

# Selecting / Filtering DataFrame by Labels

- .loc[] selects data by labels.
- Syntax: df.loc[row_labels, column_labels]
- .loc accepts:
    - Single label -> df.loc["row", "col"]
    - List of labels -> df.loc[["row1", "rown"], ["col1", "cown"]]
    - Label range -> df.loc["row1": "row3"]
    - Boolean conditions -> df.loc[df["col2"]>80, ["col1", "col2"]]

```python
df1 = pd.DataFrame({"Marks": [85,90,95], "Names":['Ankit', "Ank", "It"]}, index=["A1", "A2", "A3"])
df1.loc[["A1", "A3"]]
```

```python
df1.loc[["A1"],"Names"]
```

```python
df1.loc[["A2"], ["Names", "Marks"]]
```

```python
# Using Lambda
df1.loc[lambda x: x["Marks"]>=90]
```

```python
df1.loc[lambda x: x["Names"]=="Ank",["Names"]]
```

## Selecting / Filtering DataFrame by Position
- .iloc[] selects data by position like Python lists
- Syntax: df.iloc[row_position, col_position]
- iloc accepts:
    - Single integer df.iloc[0,1]
    - List of integers df.iloc[[0,2], [1,2]]
    - Slice ranges (end excluded) -> df.iloc[0:3, 0:2]
    - Boolean arrays -> df.iloc[[True, False, True, False],:]



```python
# Select specific index using loc
print(df1)
df1.iloc[1,1]
```

```python
# Negative Index
df1.iloc[-1:]
```

```python
df1.iloc[-1:-1]
```

```python
# Reverse order
df1.iloc[::-1]
```

```python
# Reverse row and column order
df1.iloc[::-1, ::-1]
```

## Selecting / Filtering using query
- Use .query to conditionally filter rows in a DataFrame when:
    - Need cleaner, more readable code than chained indexes
    - Need complex logical conditions involving multiple columns
    - Need to avod repetitive df[] syntax
    - Need dynamical filter based on the values of a variable

- Do not use Query in following scenarios:
    - Column names have spaces or special characters
    - When performance is critical    

```python
df1.query("Marks >= 90")
```

```python
# Apply query and then apply Column selection
df1.query("Marks >= 90")[["Names"]]
```

```python
# Query based on parameter passing and external variables
name= "Ankit"
df1.query("Names == @name")
```

```python
# Multi combination of query
name= "Ank"
df1.query("Names == @name and Marks>=90 ")
```

```python
# Parameterized query
name= "Ank"
marks = 90
df1.query("Names == @name or Marks>=@marks ")
```
