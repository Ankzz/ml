## Sorting

```python
my_list = [3,1,4,7,6,5]
sorted_list = sorted(my_list)

print(sorted_list)
```

```python
sorted_list = sorted(my_list, reverse=True)

print(sorted_list)
```

```python
my_string_list = ["advika", "arshika", "shefhali", "renu", "richa", "prachi"]
sorted_list = sorted(my_string_list, key=len)

print(sorted_list)
```

```python
def custom_sort_key(item):
    return sum(int(digit) for digit in str(item))

my_list = [13,30,22,29,36]

sorted_list = sorted(my_list, key=custom_sort_key)

print(sorted_list)
```

## Heap

Heaps are a tree implementation of a priority queues

- insert is O(log N)  
- delete.max() is O(log N)     
- heapify builds a heap in O(N)    

Defalut `heapq` module of python is a `min-heap`

To realize `max-heap` from same `heapq` store negative values of the node value.

### heapq python module

```python
import heapq

h = []
heapq.heappush(h, 3)
print(h)
heapq.heappush(h, 2)
print(h)
heapq.heappush(h,5)
print(h)
heapq.heappush(h,1)

print(h)
```

```python
minimum = heapq.heappop(h)
print(minimum)
print(h)
```

```python
minimum = heapq.heappushpop(h, 0)
print(minimum)
print(h)
```

```python
minimum = heapq.heapreplace(h, 0)
print(minimum)
print(h)
```

```python
minimum = heapq.heapreplace(h, 4)
print(minimum)
print(h)
```

```python
largest = heapq.nlargest(2, h)
print(largest)
print(h)
```

```python
smallest = heapq.nsmallest(2, h)
print(smallest)
print(h)
```

```python
import heapq
l1 = [3,2,1]
l2 = [5,0,6]
l3 = heapq.merge(l1, l2)

print(list(l3))
```

```python
heapq.heapify(l1)
heapq.heapify(l2)

l5 = heapq.merge(l1, l2)
print(list(l5))
```
