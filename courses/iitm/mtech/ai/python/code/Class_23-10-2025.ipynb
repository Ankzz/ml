## View, Copies, and Inplace Operations

```python
# If a dataframe object is assigned to another variable, the second variable gets reference to same object instance
df = pd.DataFrame({
    "Hello":[ "World!" ]
})

print(df)

df1 = df

df1["Hello"] = ["Advika!"]

print(df)
```

```python
# Creates a different object instance with copy
df1 = df.copy()

# Now this assignment does not overwrites original data-frame
df1["Hello"] = ["World!"]

print(df)
```

### Even in case Column selection, Slicing, Filtering using loc or iloc - view is returned

### Best practices:
- Selection, Slicing, Filtering using loc or iloc - view is returned
- Use copy method explicitly for creating a copy
- _is_view method to check if a variable is a copy or view


```python

df1 = df

df1._is_view
```

```python
df1 = df.copy()

df1._is_view
```

### Inplace operations in Dataframes
- Inplace operation refers to modifying an existing dataframe without returning new object
- inplace=True parameter
- Note: Does NOT happen at memory level and pandas creates a new object and reassigns it to same variable
- When to use:
  - When wiring data cleaning pipelines, original data-set not required
  - When memory is not a major concern
- When not to use:
  - When original data is not required
  - When concise code needs to be created  

```python
import numpy as np

df = pd.DataFrame({
    "Department": ["Sales", "Sales", "HR", "IT", None, "Sos"],
    "Employee": ["Asha", "Bala", "Chetan", None, "Esha", "Usha"],
    "Salary": [50,55,50,53,60, np.nan]
})

df.dropna()
df
```

## Setting / Updating Values in DataFrame

- Set value to a single cell, multiple cells at once, entire column at once
- Set or update values using:
  - Label based assignment with .loc[] or .at[]
  - Position based assignment with .iloc[] or .iat[]
  - Conditional assignment
  - Vectorized assignment
- Tips:
  - Vectorised assignment preferred
  - Label based updates preferred

**Note** : Pandas sometimes copy and sometimes views and is unpredictable in nature.  

```python
df.loc[1, "Marks"] = 65

df
```

```python
df.loc[df["Department"]=="Sales", "Salary"] += 10

df
```

```python
df1 = df.copy()

df1["Marks"] = df1["Department"].where(df1['Marks'] == None, 45)

df1
```

## Splitting Data into Groups

- groupby() operation allows:
  - Split data into groups
  - Apply a function
  - Combine results into a new DataFrame
- Aggregation functions:
  - Grouped data can be aggregated using built-in functions (mean, sum, count, min, max, etc)
  - Multiple aggregations per group
  - Custom Functions and Lambda functions

```python
df = pd.DataFrame({
    "Department": ["Sales", "Sales", "HR", "IT", "HR", "Sos"],
    "Employee": ["Asha", "Bala", "Chetan", "Rohit", "Esha", "Usha"],
    "City": ['Hyd','Del', 'Chn', 'Pat', 'Ran', 'Ahm'],
    "Experience": [1,2,3,4,5,6]
})
```

```python
grouped = df.groupby("Department")

grouped.groups
```

```python
grouped.get_group('HR')
```

```python
grouped["Experience"].agg("sum")
```

```python
grouped["Experience"].agg(["mean", "sum"])
```

```python
def range(x):
    return x.max() - x.min()

grouped["Experience"].agg(["mean", "sum", range])
```

```python
grouped.filter(lambda x: (x["Experience"]>= 2).any())
```

```python
grouped.filter(lambda x: (x["Experience"]>= 2).all())
```

## Handling Categories in DataFrames
- Special datatypes in pandas for categorical values
- Efficient for:
  - Repeated String values
  - Faster grouping, sorting, and filtering
- Define category order when comparisons are meaningful
  - Ex: Low, Medium, High

```python
df = pd.DataFrame({
    "Region": ["East", "West", "North", "South"]
})
print(df["Region"].dtype)
df["Region"] = df["Region"].astype("category")
print(df["Region"].dtype)
```

```python
df["Region"].cat.categories
```

```python
df["Region"] = df["Region"].cat.rename_categories({"East": 'E', "West": 'W', "North": 'N', "South": 'S'})
```

```python
df["Region"].cat.categories
```

```python
df["Region"] = df["Region"].cat.reorder_categories(['E', 'W', 'N', 'S'], ordered=True)
df["Region"].cat.categories
```

```python
df["Region"] = df["Region"].cat.add_categories(['NE', 'NW', 'SW', 'SE'])
df["Region"].cat.categories
```

## Handling Missing Values in DataFrames

- Missing values in Pandas are represented by:
  - NaN : Not a number
  - NaT : Not a Time
  - None: For object/string columns
- Treated as missing indicators by Pandas  
- Standard methods:
  - df.isna() & df.notna()
  - df.info()

```python
df = pd.DataFrame({
    "Name": ["Ankit", "Advika", "Arshika", "Shrija", "Kuchhum", "Puchhu", "Prasit"],
    "Marks": [np.nan, 99, 99, 97, 95, 99, 100],
    "City": [np.nan, "BGL", "BGL", "PUN", "GGN", "GGN", "PAT"]
})

df
```

```python
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

df
```
