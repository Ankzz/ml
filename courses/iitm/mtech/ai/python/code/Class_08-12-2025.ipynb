### Tensor

- Tensor is a multi-dimensional array that generalizes many other data forms
- Generalizes scalars (0D), vectors (1D), matrices (2D), and higher-dimensional arrays (3D and above)
- Tensors can hold elements of a homogeneous data type, such as integers, floats, or complex numbers
- Rank of a Tensor: No. of dimensions (axes) a tensor has
- Shape of a Tensor: Size of tensor along each axis (same as in numpy)

- Examples:
    - Scalar: A single number, rank 0: 7
    - Vector: Rank 1 and shape (n,): [1,23]
    - Matrix: Rank 2 and shape (n,2): [[1,2], [3,4]]
    - 3D Tensor: Rank 3 and shape (2, 2, 2): [[[1,2],[3,4],[4,5]],[[6,7],[7,8],[8,9]],[[[10,11],[12,13],[14,15]]]]

- Math Operations: supports a variety of math operations

**Speciality** - In terms of utilizing special hardwares - like GPU, TPU - capabilities for compute operations

#### Computational Graph
- Directed Acyclic Graph (DAG) - which is useful in performing computations in a memory and time efficient manner
    - Allows multiple parallel processing
- Nodes: Represents data (tensors) or functions/operations
- Edges: Represents data dependency or data flowing between operations

- Example: f(x,y,z) = (x+y)*z

Two types:
- Static Computation Graph
    - Known as Declarative Graph
    - Define graph structure before any computation starts
    - Allows for optimization and efficient deployment as graph is known
    - Debugging is hard because of static nature
    - Less flexible
    - Tensorflow 1.x uses this type of graph

- Dynamic Computation Graph
    - Known as Imperative Graph
    - Graph structure can be dynamically updated during execution
    - Less optimized performance
    - Debugging is easy as computations can be inspected during execution
    - More flexible
    - PyTorch and Tensorflow 2.0 use this type of graph.
